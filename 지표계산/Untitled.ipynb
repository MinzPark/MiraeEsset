```python
from pykrx import stock
from datetime import date
import pandas as pd
```

```python
def daily_rtn(df):
    temp = pd.DataFrame(df['종가'].pct_change(1)).rename(columns = {'종가':'daily_rtn'})
    return temp
```

```python
def month_rtn(df,year): # input: [종가 df,year_list]
    for k,yy in enumerate(year):

            # 각 년도별 달 
            each_mth = df.loc['{}'.format(yy)].index.month.unique().tolist()
            for idx,i in enumerate(each_mth):
                # 30일인 달
                if(i<10 and i%2!=0 and i!=9)|(i ==8 or i == 10 or i ==12):
                    if i<10:
                        mm = '0'+str(i)
                    else:
                        mm = str(i)

                    month_stock = df.loc['{}-{}-01'.format(yy,mm):'{}-{}-31'.format(yy,mm)]
                    month_trading = month_stock['종가'].pct_change(periods = len(month_stock)-1)[-1]
                    rtn = [month_trading]*len(month_stock)
                    month_rtn = pd.DataFrame(data = rtn, index = month_stock.index,columns = ['month_rtn'])
                # 2월
                elif (i ==2):
                    mm = '0'+str(i)
                    month_stock = df.loc['{}-{}-01'.format(yy,mm):'{}-{}-28'.format(yy,mm)]
                    month_trading = month_stock['종가'].pct_change(periods = len(month_stock)-1)[-1]
                    rtn = [month_trading]*len(month_stock)
                    month_rtn = pd.DataFrame(data = rtn, index = month_stock.index,columns = ['month_rtn'])

                # 31일인 달
                else:
                    if i<10:
                        mm = '0'+str(i)
                    else:
                        mm = str(i)
                    month_stock = df.loc['{}-{}-01'.format(yy,mm):'{}-{}-30'.format(yy,mm)]
                    month_trading = month_stock['종가'].pct_change(periods = len(month_stock)-1)[-1]
                    rtn = [month_trading]*len(month_stock)
                    month_rtn = pd.DataFrame(data = rtn, index = month_stock.index,columns = ['month_rtn'])


                if idx==0:
                    total = month_rtn

                else:
                    total = pd.concat([total,month_rtn])

            if k == 0:
                total_month_rtn = total
            else:
                total_month_rtn = pd.concat([total_month_rtn,total])
    return total_month_rtn
            
```
