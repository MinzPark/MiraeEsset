```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
def rfr():
    # 값 추출
    url = 'https://www.kofr.kr/main.jsp'
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(3)

    source = driver.page_source
    html = BeautifulSoup(source)
    rfr = html.find('strong',class_ = 'w2textbox textXl').get_text()
    return float(rfr/100)
```

```python
rfr = float(rfr()) # 백분율값이므로 필요시 값 /100처리!
type(float(rfr))
```
