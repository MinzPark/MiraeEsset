```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from tqdm import tqdm 
```

```python
def last_news_content_pg(code):
    driver = webdriver.Chrome()
    driver.implicitly_wait(10) # 페이지가 다 뜰때까지 최대 10

    # 네이버 금융 뉴스공시탭
    driver.get("https://finance.naver.com/item/news.naver?code={}".format(code))
    time.sleep(3)

    # 내용 기준 제목 
    driver.switch_to.frame('news_frame')
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '''/html/body/div/div/div[1]/a[2]'''))).click()
    
    # newsframe 접근 뉴스의 맨끝 번호 (for문의 제한양)
    driver.switch_to.frame('news_frame')
    source = driver.page_source
    html = BeautifulSoup(source)
    last_pg = int(html.find_all('a')[-1].get('href').split('=')[2].split('&')[0])
    driver.quit()
    
    return last_pg
# [정수형 마지막 페이지]
```

```python
def news_content_title(code):    
    total = []
    
    # 뉴스탭의 마지막 페이지 가져오기 
    last_pg = last_news_content_pg(code)
    
    # 네이버 금융 뉴스공시탭    
    driver = webdriver.Chrome()
    driver.implicitly_wait(10) # 페이지가 다 뜰때까지 최대 10

    driver.get("https://finance.naver.com/item/news.naver?code={}".format(code))
    time.sleep(3)
    
    # 내용 기준 제목 
    driver.switch_to.frame('news_frame')
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '''/html/body/div/div/div[1]/a[2]'''))).click()
    
    for i in range(1, int(last_pg/10)+1):
        print('{}~{} 출력중'.format((i-1)*10,(i-1)*10+10))
        if i == 1:
            for j in tqdm(range(1,13)):
                if j ==3: # 2쪽이 중복됨 이를 제거 i.e, '''/html/body/div/table[2]/tbody/tr/td[{}]/a'''.format(2)|format(3)값이 같음 
                    continue
                else:
                    driver.switch_to.default_content()
                    driver.switch_to.frame('news_frame')
                    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '''/html/body/div/table[2]/tbody/tr/td[{}]/a'''.format(j)))).click()
                    source = driver.page_source
                    html = BeautifulSoup(source)
                    title = [item.find('a').get_text() for item in html.find_all('td',class_ = 'title')]
                    total.append(title)
        else:
            for j in tqdm(range(4,14)):
                driver.switch_to.default_content()
                driver.switch_to.frame('news_frame')
                WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '''/html/body/div/table[2]/tbody/tr/td[{}]/a'''.format(j)))).click()
                source = driver.page_source
                html = BeautifulSoup(source)
                title = [item.find('a').get_text() for item in html.find_all('td',class_ = 'title')]
                total.append(title)
                
                if total[-2] == total[-1]: # 마지막이 삼성의 경우 1267여도 모든 게 새로운 제목이 아닌 반복된 제목이 채워진 것일 수 있음. 
                    break

        if total[-2] == total[-1]: # 마지막이 삼성의 경우 1267여도 모든 게 새로운 제목이 아닌 반복된 제목이 채워진 것일 수 있음. 
                break
    
    total.pop() # 겹친거 체크한 후 제거 
    
    return total
```
